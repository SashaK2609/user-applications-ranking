```python
import pandas as pd

file_path = '/content/drive/My Drive/Colab Notebooks/applications(2.0).csv'
applications_df = pd.read_csv(file_path)
applications_df
```

```python
applications_df = applications_df.drop_duplicates(subset = ['applicant_id'], keep = 'first')
print(applications_df)
```

```python
applications_df['External Rating'] = applications_df['External Rating'].fillna(0)
print(applications_df.head())
```

```python
applications_df['Education level'] = applications_df.loc[:,('Education level')].fillna('Середня')
print(applications_df.head(50))
```

```python
file_path_2 = '/content/drive/My Drive/Colab Notebooks/industries(2.0).csv'
industries_df = pd.read_csv(file_path_2)
industries_df
```

```python
merged_df = pd.merge(applications_df, industries_df, on = 'Industry', how = 'left')
print(merged_df)
```

Рейтинг

```python
# age is between 35 and 55 add 20 points
merged_df['Rating_Age'] = merged_df['Age'].apply(lambda x: 20 if 35 <= x <= 55 else 0)
print(merged_df['Rating_Age'].astype(int))
```

```python
from datetime import datetime
merged_df['Applied at'] = pd.to_datetime(merged_df['Applied at'], format = 'mixed')
print(merged_df['Applied at'])
```

```python
day_of_the_week = merged_df['Applied at'].dt.day_name()
print(day_of_the_week)
```

```python
# applied on weekdays add 20 points
merged_df['rating_day'] = day_of_the_week.apply(lambda day: 20 if day in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'] else 0)
print(merged_df['rating_day'])
```

```python
#applicant is married add 20 points
merged_df['rating_marital_status'] = merged_df['Marital status'].apply(lambda x: 20 if x == 'Married' else 0)
print(merged_df['rating_marital_status'])
```

```python
#applicant located in Kyiv or region add 10 points
merged_df['rating_location'] = merged_df['Location'].apply(lambda x: 10 if x == 'Київ чи область' else 0)
print(merged_df['rating_location'])
```

```python
merged_df['rating_7'] = merged_df['External Rating'].astype(int).apply(lambda x: x + 20 if x >= 7 else x)
print(merged_df['rating_7'])
```

```python
merged_df['rating_sum'] = pd.Series([merged_df['Rating_Age'], merged_df['rating_day'], merged_df['rating_marital_status'], merged_df['rating_location'], merged_df['rating_7'], merged_df['Score']]).sum()
print(merged_df['rating_sum'])
```

```python
def rating_function(row):
    if row['External Rating'] <= 2:
        return row['rating_sum'] - 20
    elif pd.isnull(row['Amount']) or row['External Rating'] == 0:
        return 0
    else:
        return row['rating_sum']
```

```python
merged_df['Rating'] = merged_df.apply(rating_function, axis=1)
print(merged_df['Rating'])
```

```python
#applications with rating more than 0
merged_df = merged_df[merged_df['Rating'] != 0]
print(merged_df['Rating'].reset_index(drop=True))
```

```python
week_rating = merged_df.groupby(merged_df['Applied at'].dt.isocalendar().week)['Rating'].mean()
week_rating
```

```python
week_rating.plot(kind = 'bar')
```
